### GRADIENT DESCENT

```python
import numpy as np
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])

def grad_desc(x,y):
    iterations = 166
    m_curr = b_curr = 0
    n = len(x)
    learn_rate = 0.08
    for i in range (0,iterations):
        y_predicted = m_curr*x + b_curr
        #doubt here:
        cost_mse = (1/n) * sum([val**2 for val in (y-y_predicted)])
        md = -(2/n)*sum(x*(y-y_predicted))
        bd = -(2/n)*sum(y-y_predicted)
        m_curr = m_curr - learn_rate*md
        b_curr = b_curr - learn_rate*bd
        print(f"current m :{m_curr} , current b :{b_curr}, cost_fn_mse = {cost_mse} , iteration = {i} ")

grad_desc(x,y)

```
